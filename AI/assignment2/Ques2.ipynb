```python
#simulated Annealing

def display_board(state):
    n = len(state)
    for row in range(n):
        line = ""
        for col in range(n):
            if state[col] == row:
                line += " Q "
            else:
                line += " . "
        print(line)
    print("-" * (n * 3))


def calc_conflicts(state):
    conflicts=0
    for i in range(len(state)):
        for j in range(i+1, len(state)):
            if state[i] == state[j] or abs(i-j) == abs(state[i] - state[j]):
                conflicts += 1
    return conflicts


def n_queens_simulated_annealing(state, temp, cooling_rate):
    current = state
    display_board(state)
    random.shuffle(current)
    current_conflicts = calc_conflicts(current)

    iteration = 1


    while iteration <= 20000:

        if current_conflicts == 0:
            break
        
        new_state = current.copy()
        print("Current State:")
        display_board(new_state)
        
        n = len(current)
        neighbor = current[:]
        col = random.randint(0, n - 1)
        neighbor[col] = random.randint(0, n - 1)
        new_state = neighbor

        new_conflicts = calc_conflicts(new_state)
        print(f"New Conflicts: {new_conflicts}")

        delta = current_conflicts - new_conflicts
        print(f"Delta: {delta}")

        if delta > 0 or (temp > 0 and random.random() < np.exp(-delta/temp)):
            print("Accepting worst state:")
            current = new_state
            current_conflicts = new_conflicts

        print(f"Iteration: {iteration}, Temp: {temp: .6f}, Energy/Conflicts: {current_conflicts}")
        display_board(current)

        temp *= cooling_rate
        iteration += 1
    return current, current_conflicts, iteration
```

```python
print("Starting configuration:")
initial_state = [3, 0, 0, 5, 2, 2, 1, 4] #[6, 3, 4, 2, 0, 7, 5, 1]
display_board(initial_state)

print("Solution using Simmuleated annealing: ")

solution, energy, steps = n_queens_simulated_annealing(initial_state, temp=1.0, cooling_rate=0.95)
print(f"\nSolution found in {steps} steps")
print(f"Final energy: {energy}")
print("Final solution state/board")
display_board(solution)
```
