### Question - 2

```python
from sklearn.impute import SimpleImputer
# Filling missing values

def fillMissingValues(df, strategy = "mean"):

    # Separate numerical and categorical columns
    numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
    categorical_cols = df.select_dtypes(include=['object']).columns

    # Impute numerical columns
    num_imputer = SimpleImputer(strategy=strategy)
    df[numerical_cols] = num_imputer.fit_transform(df[numerical_cols])

     # Impute categorical columns
    cat_imputer = SimpleImputer(strategy='most_frequent')
    df[categorical_cols] = cat_imputer.fit_transform(df[categorical_cols])
    
    return df
```

### Question - 4

```python
from sklearn.preprocessing import OneHotEncoder

def encode_categorical_values(df):

    df_encoded = df.copy()
    categorical_cols = df_encoded.select_dtypes(include=['object']).columns.tolist()

    # Initialize the encoder
    encoder = OneHotEncoder(sparse_output = False)

    encoded = encoder.fit_transform(df_encoded[categorical_cols])

    #We use get_feature_names_out() to get the column names for the encoded data
    one_hot_encoded_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out(categorical_columns))

    # Concatenate the one-hot encoded dataframe with the original dataframe
    df_encoded = pd.concat([df_encoded, one_hot_encoded_df], axis=1)

    # Drop the original categorical columns
    df_encoded = df_encoded.drop(categorical_columns, axis=1)
    return df_encoded
    
```

### Question - 5

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

def normalize(df):
    # creating a copy of the dataset
    df_normalize = df.copy()
    #selecting numerical columns for normalizing
    numerical_col = df.select_dtypes(include=['int64','float64']).columns
    
    scaler = MinMaxScaler()
    # Used MinMaxScaler() for normalizing the numerical column
    df_normalize[numerical_col] = scaler.fit_transform(df_normalize[numerical_col])
    
    return df_normalize
```

### Question - 6

```python
from sklearn.model_selection import train_test_split
def split_data(df):
    X = df.drop('Max. Price', axis=1)
    Y = df['Max. Price']

    
    # split out the train set as 70% and test+val set as 30%
    x_train, x_temp, y_train, y_temp = (X, Y, test_size = 0.3, random_state = 42)

    #splitting the test and val set
    x_test, x_val, y_test, y_val = (X, Y, test_size=0.333, random_state=42)

    return x_train, x_test, x_val, y_train, y_test, y_val    
```
